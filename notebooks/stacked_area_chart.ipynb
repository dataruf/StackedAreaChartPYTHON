### Lazım olan kitabxanalar: Pandas, Matplotlib, Seaborn.

### Analiz üçün Gas, Petrol və Diesel yanacaq növlərinin saxlanması. (other elektrikli ve s. ola bilər)

```python
df_fuel_type = df[['Year', 'Gas', 'Petrol', 'Diesel']]
```

```python
df_fuel_type.head()
```

Yeni datasetimi df_fuel_type (analiz edəcəyimiz)

## Stacked-Area chartının qurulması.

```python
colors = ["#171C63", "#BC3EB4", "#FF0404"]
```

```python
sns.set_style('white')
```

```python
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_type['Year'],
              df_fuel_type['Diesel'],
              df_fuel_type['Petrol'],
              df_fuel_type['Gas'],
              colors= colors,
              edgecolor = 'black')
plt.legend(['Diesel', 'Petrol', 'Gas'], loc = 'upper right')
plt.xticks(df_fuel_type['Year']
, rotation = 90) #45 >>>>> 90
plt.ylabel('Number of cars', fontsize = 15)
plt.title('Engine type popularity (1982-2016)',
          fontsize = 20, weight = 'bold',
          loc = 'center')
sns.despine()
plt.savefig('stacked_area_chart.png')
plt.show()
```
